```python
#Import Modules, Libraries, Etc. for EDA
import pandas as pd
import numpy as np
import json
```

```python
trump = pd.read_json("PresidentTweets.json")
```

**Some Cleaning/Checking**

```python
#Tells us non-null types
trump.info()
```

```python
#types of data
trump.describe()
```

```python
#id_str is meaningless, so we're getting rid of it. 
trump = trump.drop(["id_str"], axis = 1)
```
